### 匯入所需套件

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

### 輸入資料

```python
# 輸入資料
dir_data = '/Users/a7890123119/Downloads'
f_app = os.path.join(dir_data, 'train.csv')
app_train = pd.read_csv(f_app)
app_train
```

### 繪製長條圖

```python

# 將 MSSubClass 轉換為字串
app_train["SalePrice"]= app_train["SalePrice"].astype("str")

# 計算 MSSubClass 的數量
class_counts = app_train["SalePrice"].value_counts()


# 繪製長條圖
plt.bar(x=class_counts.index, 
        height=class_counts.values)

# 添加圖表標題及標籤
plt.title("Area House Price")
plt.xlabel("Contribution")
plt.ylabel("Price")

# 顯示圖表
plt.show()
```

### 繪製盒鬚圖

```python
# 輸入資料
raw_data = pd.read_csv("train.csv")

plt.boxplot(raw_data["SalePrice"])
plt.show()
```

### 繪製散佈圖

```python
# 輸入資料
raw_data = pd.read_csv("train.csv")

# 探索每個欄位資訊
raw_data.info()

# 計算年齡與食物消費金額之間的關係
### 第一種寫法 ###
plt.scatter(x = raw_data["SalePrice"], 
            y = raw_data["YrSold"])

### 第二種寫法 ###
plt.scatter(x = "SalePrice",
            y = "YrSold",
            data = raw_data)

plt.show()
```

### 繪製折線圖

```python
raw_data = pd.read_csv("train.csv")


raw_data["Sale_Group"] = pd.cut(raw_data["SalePrice"], 5, labels = ["Sale-1", "Sale-2", "Sale-3", "Sale-4", "Sale-5"])

plt.plot(raw_data["Sale_Group"], raw_data["SalePrice"])
plt.show()
```
