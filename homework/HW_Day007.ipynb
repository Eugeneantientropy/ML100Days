## 匯入套件

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

## 匯入資料與簡單的資料探索

```python
# 輸入資料
dir_data = 'C:/Users/a7890/Downloads/'
f_app = os.path.join(dir_data, 'train.csv')
raw_data = pd.read_csv(f_app)
raw_data


# 探索每個資料欄位與遺失值概況
```

```python
# 為求學習方便，在此先將有遺失值的資料剔除掉
```

```python
num_of_transported = raw_data['SalePrice'].value_counts()
```

## 使用 Subplot 建立子圖


```python
fig = plt.figure()
plt.subplot(2, 2, 1)
plt.bar(x=num_of_transported.index,
        height=num_of_transported.values)

plt.subplot(2, 2, 2)
plt.boxplot(raw_data['SalePrice'])

plt.subplot(2, 2, 3)
plt.scatter(raw_data.index, raw_data['SalePrice'])

plt.subplot(2, 2, 4)
plt.plot(num_of_transported.index, num_of_transported.values)
```

##  使用 Subplots 建立子圖

```python
# Reset fig

fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0][0].bar(x=num_of_transported.index, height=num_of_transported.values)
ax[0][1].boxplot(raw_data['SalePrice'])
ax[1][0].scatter(raw_data.index, raw_data['SalePrice'])
ax[1][1].plot(num_of_transported.index, num_of_transported.values)

plt.show()
```

## 使用 add_subplot 建立子圖



```python

fig = plt.figure()
plt.subplot(2, 2, 1)
plt.bar(x=num_of_transported.index,
        height=num_of_transported.values)

plt.subplot(2, 2, 2)
plt.boxplot(raw_data['SalePrice'])

plt.subplot(2, 2, 3)
plt.scatter(raw_data.index, raw_data['SalePrice'])

plt.subplot(2, 2, 4)
plt.plot(num_of_transported.index, num_of_transported.values)

plt.show()
```

## 使用 GridSpec 建置不規則排列

```python
raw_data = pd.read_csv(f_app)

fig = plt.figure()
grid = plt.GridSpec(5, 4, wspace = 0.2, hspace = 0.4)

main_scatter = plt.subplot(grid[:3, 1:]) # 決定散佈圖會包含哪些子圖
main_scatter.scatter(x = "MSSubClass", y = "SalePrice", data = raw_data)
main_scatter.set_xticklabels("") # 將散佈圖的 X 軸上的數字刪除掉
main_scatter.set_yticklabels("") # 將散佈圖的 Y 軸上的數字刪除掉

y_hist = plt.subplot(grid[:3, 0])
y_hist.hist(x = "SalePrice", data = raw_data, orientation = "horizontal")
y_hist.invert_xaxis() # 將直方圖結果由左邊翻轉至右邊
y_hist.set_ylabel("SalePrice")

x_hist = plt.subplot(grid[3:, 1:])
x_hist.hist(x = "SalePrice", data = raw_data)
x_hist.invert_yaxis() # 將直方圖結果由下方翻轉至上方
x_hist.set_xlabel("SalePrice")

fig.tight_layout()
plt.show()
```
